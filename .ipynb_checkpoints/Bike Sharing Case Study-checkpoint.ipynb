## Training the model

### Feature selection
-  manually removing features based on p values and VIF
-  using automated approach
-  using a combinatioin of both


```python
# rescaling of train data
scaler = MinMaxScaler()
var_list =['temp','atemp','hum','windspeed','casual','registered','cnt']
df_train[var_list]=scaler.fit_transform(df_train[var_list])
df.head()
```

```python
# rescaling of test data
var_list =['temp','atemp','hum','windspeed','casual','registered','cnt']
df_test[var_list]=scaler.transform(df_test[var_list])
df.head()
```

```python
# plot the correlation between variables
plt.figure(figsize=(12,9))
res = df_train.corr()
sns.heatmap(res,annot=True)
plt.show()

# target variable cnt has high correlation with registered,casual,yr, and temp
```

```python
y_train = df_train.pop('cnt')
X_train = df_train

y_test = df_test.pop('cnt')
X_test = df_test
```

```python
# drop casual and registered because cnt is combination of both.So adding them in the model will lead to incorrect results.
# Casual and regiseterd are linearly related
X_train.drop(['casual','registered'],inplace=True,axis=1)
```

#### Approach 1

```python
# model with all feature variables
X_train_sm = sm.add_constant(X_train)
lr = sm.OLS(y_train,X_train_sm)

lr_model = lr.fit()
lr_model.summary()

# r_squared is 0..834.
# p value of mnth and atemp is really high
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif_df = pd.DataFrame()

vif_df['Features']=X_train_sm.columns
vif_df['VIF']= [variance_inflation_factor(X_train_sm.values,i) for i in range(X_train_sm.shape[1])]
vif_df

## variables temp and atemp have high VIF. There mostly alike and have a strong correlation. So one of the,m has to be removed
```

```python
# drop atemp as it has highest VIF
X_train_sm.drop('atemp',axis=1,inplace=True)
```

```python
# model with year variables
lr = sm.OLS(y_train,X_train_sm)

lr_model = lr.fit()
lr_model.summary()

# month has really high p value 
```

```python
vif_df = pd.DataFrame()

vif_df['Features']=X_train_sm.columns
vif_df['VIF']= [variance_inflation_factor(X_train_sm.values,i) for i in range(X_train_sm.shape[1])]
vif_df

# mnth has to be dropped because it has high p value
```

```python
X_train_sm.drop('mnth',axis=1,inplace=True)
```

```python
# model with year variables
lr = sm.OLS(y_train,X_train_sm)

lr_model = lr.fit()
lr_model.summary()
```

```python
vif_df = pd.DataFrame()

vif_df['Features']=X_train_sm.columns
vif_df['VIF']= [variance_inflation_factor(X_train_sm.values,i) for i in range(X_train_sm.shape[1])]
vif_df

# the values of VIF are in acceptable range. So the model is good.
```

```python
y_train_pred=lr_model.predict(X_train_sm)
res = y_train - y_train_pred
sns.distplot(res)
plt.title("Residual error distribution")
plt.show()
```

```python
# prediction
drop_var_list = ['casual','registered','atemp','mnth']
X_test.drop(drop_var_list,axis=1,inplace=True)
X_test_sm = sm.add_constant(X_test)
```

```python
y_test_pred = lr_model.predict(X_test_sm)
```

```python
from sklearn.metrics import r2_score
r_squared = r2_score(y_true=y_test,y_pred=y_test_pred)
r_squared
## r_squared for the first model is 0.80, which is pretty good.
```
